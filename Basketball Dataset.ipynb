```python
def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c = 'black',ls = '--',marker='o',ms =7,label = name)
    plt.xticks(list(range(0,9)),Seasons,rotation = 'vertical')
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.show()
```

```python
#Advanced function
def myplot(data,playerlist = Players):
    Col = {"KobeBryant":"Black","JoeJohnson":"red","LeBronJames":"blue","CarmeloAnthony":"magenta","DwightHoward":"green","ChrisBosh":"violet","ChrisPaul":"yellow","KevinDurant":"purple","DerrickRose":"green","DwayneWade":"red"}
    Mar = {"KobeBryant":"s","JoeJohnson":".","LeBronJames":"o","CarmeloAnthony":"^","DwightHoward":".","ChrisBosh":"s","ChrisPaul":"o","KevinDurant":".","DerrickRose":"^","DwayneWade":"s"}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c = Col[name],ls = '--',marker= Mar[name],ms =7,label = name)
    plt.xticks(list(range(0,9)),Seasons,rotation = 'vertical')
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.show()
```
